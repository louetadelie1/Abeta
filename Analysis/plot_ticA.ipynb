### PLOTTING FIRST 1000 TS

```python
colvar=np.genfromtxt("/pool/work/adelie/results_G5_der/cv_compare/ligandD/COLVAR_reg123")
time=colvar[:1000,0]
reg1=colvar[:1000,1]
reg2=colvar[:1000,2]
reg3=colvar[:1000,3]

y = time
fig = plt.figure()
ax1 = fig.add_subplot(111)

ax1.scatter(reg1, y, s=10, c='cornflowerblue', marker="s", label='reg1')
ax1.scatter(reg2,y, s=10, c='orange', marker="o", label='reg2')
ax1.scatter(reg3,y, s=10, c='crimson', marker="*", label='reg3')
plt.ylabel("Simulation Time (ts)")
plt.xlabel("Frequency")
plt.legend(loc = 'upper right')
plt.title("Coordination Number")

plt.show()
```
